### Label Encoding

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
columns = ['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','OverTime']
data[columns]=data[columns].apply(le.fit_transform)
```

```python
data.head()
```

#### Splitting Dependent and Independent Variables

```python
x = data.drop(columns='Attrition')
y = data['Attrition']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
x_scaled=pd.DataFrame(ms.fit_transform(x),columns=x.columns)
```

#### Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size = 0.2,random_state = 0)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

## Model Building

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

### Evaluation of the model 

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score,roc_curve
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
pd.crosstab(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
probability=lr.predict_proba(x_test)[:,1]
```

```python
y_test = y_test.map({'Yes': 1, 'No': 0}).astype(int)
fpr,tpr,threshsholds = roc_curve(y_test,probability)
```

```python
plt.plot(fpr,tpr)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC CURVE')
plt.show()
```
