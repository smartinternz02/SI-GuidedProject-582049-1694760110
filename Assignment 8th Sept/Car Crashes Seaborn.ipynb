### Importing Car crashes dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
car_crashes = sns.load_dataset("car_crashes")
```

```python
#Getting an overview of the dataset
car_crashes.head()
```

```python
#Checking basic statistics of the variables
car_crashes.describe()
```

### Data Visualization

```python
sns.pairplot(car_crashes)
plt.title("Car Crashes Data")
plt.show()
```

Inference: The pairplot provides a matrix of scatterplots for each pair of numerical variables. It helps visualize relationships and distributions between variables. For example, there seems to be  a positive correlation between variables like "total" and "alcohol," suggesting that more alcohol-related crashes tend to occur in areas with higher total crashes

```python
sns.histplot(car_crashes["total"], kde=True)
plt.title("Total Crashes")
plt.xlabel("Total Crashes")
plt.ylabel("Frequency")
plt.show()
```

Inference: The histogram of total crashes shows the distribution of total crash counts. It appears to be slightly right-skewed, indicating that most areas have a lower number of crashes, with a few areas having relatively higher crash counts.

```python
correlation_matrix = car_crashes.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Heatmap of Correlation")
plt.show()
```

Inference: The heatmap of correlation displays the correlation coefficients between numerical variables. Darker colors represent stronger correlations. For instance, "alcohol" and "total" have a strong positive correlation, indicating that areas with more alcohol-related crashes tend to have more total crashes.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y="alcohol", data=car_crashes)
plt.title("Box Plot of Alcohol-Related Crashes")
plt.ylabel("Alcohol-Related Crashes")
plt.show()
```

Inference: The box plot provides insights into the distribution of alcohol-related crashes across different areas. It shows the central tendency, spread, and any potential outliers in the data related to alcohol-related crashes.
